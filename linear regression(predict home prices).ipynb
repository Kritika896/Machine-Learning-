```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with columns 'area_sqft' and 'price_usd'
# Example DataFrame creation
data = {'area_sqft': [2600, 3000, 3200, 3600,4000],
        'price_usd': [550000,565000 ,610000 ,680000,725000 ]}
df = pd.DataFrame(data)

# Plotting
plt.scatter(df['area_sqft'], df['price_usd'], color='red', marker='+')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (US$)')
plt.title('Scatter plot of Price vs Area')
plt.show()


```

```python
from sklearn import linear_model
data = {'area_sqft': [2600, 3000, 3200, 3600,4000],
        'price_usd': [550000,565000 ,610000 ,680000,725000 ]}
df = pd.DataFrame(data)

# Creating a linear regression model
reg = linear_model.LinearRegression()

# Fitting the model
reg.fit(df[['area_sqft']], df['price_usd'])
```

```python
prediction = reg.predict([[3300]])
prediction
```

```python
#value of slope
reg.coef_
```

```python
#value of intercept 
reg.intercept_
```
